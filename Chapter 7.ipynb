# 准备数据（手动生成数据）

```python
def generate(samples_size,mean,cov,diff,one_hot):
    num_classes = 2
    samples_per_classes = int(samples_size/num_classes)            # 控制样本分布：每个类别对应的样本数目基本平衡
    
    x0 = np.random.multivariate_normal(mean,cov,samples_per_classes)
    y0 = np.zeros(samples_per_classes)
    
    for ci, d in enumerate(diff):
        x1 = np.random.multivariate_normal(mean+d,cov,samples_per_classes)
        y1 = (ci+1)*np.ones(samples_per_classes)
        
        X0 = np.concatenate((x0,x1))
        Y0 = np.concatenate((y0,y1))
        
    if one_hot == True:    # one-hot 编码, pandas.get_dummies()可快速进行 one_hot 编码
        class_ind = [Y0==class_number for class_number in range(num_classes)]
        Y0 = np.asarray(np.concatenate((np.array(class_ind[0]).reshape((-1,1)),np.array(class_ind[1]).reshape((-1,1))),1))
    
#     data = np.array(list(zip(X0,Y0)))
    data = list(zip(X0,Y0))
    np.random.shuffle(data)
    
    return data
```
